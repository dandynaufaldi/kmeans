```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from kmeans import KMeans
import utils
%matplotlib inline
```

```python
data = pd.read_csv('seed.txt', sep='\t', header=None)
```

```python
data.head()
```

```python
x = data.drop([7], axis=1)
y = data[7]
```

```python
x.describe()
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
x_normalized = utils.normalize(x.values)
```

```python
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(x_normalized)
```

```python
x_pca.shape
```

```python
y.loc[y==1].index
```

```python
y.unique()
```

```python
colors = ['red', 'blue', 'green']
for i, cls in enumerate(y.unique()):
    idx = y.loc[y == cls].index
    plt.scatter(x_pca[idx, 0], x_pca[idx, 1], c=colors[i], label=cls)
# plt.scatter(x_pca[:,0], x_pca[:,1])
plt.title('Original Class')
plt.legend()
plt.savefig('docs/original_class.png')
plt.show()
```

```python
kmeans = KMeans(n_cluster=3, seed=2018)
```

```python
kmeans.fit(x_normalized)
```

```python
predicted = kmeans.predict(x_normalized)
```

```python
np.unique(predicted)
```

```python
colors = ['green', 'blue', 'red', 'orange', 'purple', 'gray']
for i, cls in enumerate(np.unique(predicted)):
    idx = np.where(predicted == cls)
    plt.scatter(x_pca[idx, 0], x_pca[idx, 1], c=colors[i], label=cls+1)
# plt.scatter(x_pca[:,0], x_pca[:,1])
plt.title('Cluster Result')
plt.legend()
plt.savefig('docs/clustered_result.png')
plt.show()
```

```python
n_clusters = list(range(2,7))
silhouettes = []
sses = []
colors = ['green', 'blue', 'red', 'orange', 'purple', 'gray']
for k in range(2, 7):
    kmeans = KMeans(n_cluster=k, seed=2018)
    kmeans.fit(x_normalized)
    label = kmeans.predict(x_normalized)
    silhouette = silhouette_score(x_normalized, label)
    silhouettes.append(silhouette)
    fig, ax = plt.subplots()
    for i, cls in enumerate(np.unique(label)):
        idx = np.where(label == cls)
        ax.scatter(x_pca[idx, 0], x_pca[idx, 1], c=colors[i], label=cls+1)
    # plt.scatter(x_pca[:,0], x_pca[:,1])
    ax.set(title='Cluster Result K = {}'.format(k))
#     ax.title('Cluster Result K = {}'.format(k))
    ax.legend()
    fig.savefig('docs/clustered_result_k{}.png'.format(k))
#     plt.show()
    
    sses.append(kmeans.SSE)
```

```python
result = pd.DataFrame()
```

```python
result['n_cluster'] = n_clusters
result['Silhouette'] = silhouettes
result['SSE'] = sses
```

```python
plt.plot(result.Silhouette)
plt.xlabel('Num of Cluster')
plt.ylabel('Silhouette')
plt.xticks(list(range(5)), result.n_cluster.values)
plt.title('Silhouette Score')
plt.savefig('docs/silhouette_score.png')
plt.show()
```

```python
plt.plot(result.SSE)
plt.xlabel('Num of Cluster')
plt.ylabel('SSE')
plt.xticks(list(range(5)), result.n_cluster.values)
plt.title('SSE Score')
plt.savefig('docs/SSE_score.png')
plt.show()
```
